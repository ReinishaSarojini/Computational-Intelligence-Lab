```python
#importing the necessary libraries
import pandas as pd
import random
import numpy as np
import array as array
import math
from statistics import mode
from prettytable import PrettyTable
```

```python
#Reading data from excel
data = pd.read_csv('/content/Iris.csv')

#Printing data
data
```

```python
#For training data,select 80 records 

#For getting features for training, drop Species column
train_x = pd.DataFrame(data.drop('Species',axis=1)[0:80])
train_y = data[0:80]

#train_y contains the target variable for the selected 80 records.
train_y = train_y['Species']

#For test case , randomly chose one record in the remaining
x=random.randrange(81,100)
test_x = data.iloc[x]
test_y =[ test_x['Species']]
test_x =test_x.drop('Species')
```

```python
#choosing differnet K values and compare the output
k =[3,5,7,9,11]

test=[]

#Find the Euclidean distance for the test data corresponding to all training data.
for j in range(len(train_x)):
  ans =0
  for i in range(4):
    ans +=(train_x.iat[j,i]-test_x.iat[i])**2
  test.append(math.sqrt(ans))

#Creating a dataframe with euclidean distance and target variable for the test data corresponding to all training data
df = pd.DataFrame({"Euclidean Distance":test,"Category":train_y})
df = pd.concat([train_x,df],axis =1)

# Adding rank to the dataframe
df['Rank'] =df["Euclidean Distance"].rank().astype(int)

#Printing the results
print("Test Data \n",test_x)
print()

#Creating a table using PrettyTable with the necessary columns
myTable = PrettyTable(["K","Nearest Neighbours","Predicted Label"])

#Adding rows to the table
for i in k:
  myTable.add_row([i,df[df.Rank<=i].iloc[:, [1,2,3,4,6]] ,mode(df[df.Rank<=i].Category)])
  
print(myTable)
```

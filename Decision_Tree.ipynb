```python
#Importing neccesary libraries
import pandas as pd
import math
```

```python
#Reading data
data = pd.read_csv('/content/PlayTennis.csv')
data
```

```python
def entropy(df):
  yes_count = len(df[df['Play Tennis']=='Yes'])
  no_count = len(df) - yes_count
  #For avoiding math domain error
  if yes_count==0 or no_count == 0:
    return 0
  return (-((yes_count/len(df))*math.log2(yes_count/len(df))) - ((no_count/len(df))*math.log2(no_count/len(df))))
```

```python
result = {}

for i in range(len(data.columns)-1):
  #For extracting unique values in particular column
  attribute_list = list(set(data.iloc[:,i]))
  temp =0
  print(list(data.columns)[i])
  for j in attribute_list:    
    df = data[data.iloc[:,i]==j]
    print(j,"(",len(df[df['Play Tennis']=='Yes']),",",len(df)-len(df[df['Play Tennis']=='Yes']),")")
    entropy_value = round(entropy(df),3)
    print("Entropy :",entropy_value,"\n")
    temp += (len(df)/len(data)) * entropy_value
  result[list(data.columns.values)[i]] = temp
  print("Entropy for ",list(data.columns.values)[i]," :",round(temp,3))
  print()

info_gain = round(entropy(data),3)
print("Entropy for whole dataset :",info_gain)
print()

print("Result \nInformation Gain = Entropy before split - Entropy after split\n")
for i in range(len(data.columns)-1):
  print("Information Gain of ",list(data.columns.values)[i]," ",round(info_gain-result[list(data.columns.values)[i]],3))

print("The Root Node is ",min(result , key = lambda x:result[x]))
```
